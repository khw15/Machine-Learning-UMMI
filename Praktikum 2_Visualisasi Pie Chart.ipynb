Visualizing data with pie charts

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
flavors = ('Chocolate', 'Vanilla', 'Pistachio', 'Mango', 'Strawberry')
votes = (12, 11, 4, 8, 7)
colors = ('#ff9999','#66b3ff','#99ff99','#ffcc99')
explode = (0, 0, 0, 0.1, 0)
```

```python
plt.title('Favorite Ice Cream Flavors')
plt.pie(
        votes[:len(flavors)],
        labels=flavors,
        autopct='%1.1f%%',
        colors=colors,
        explode=explode,
        shadow=True,
        )
plt.show()
```

Visualizing data with bar charts

```python
countries = ('Malaysia', 'Singapore', 'Indonesia', 'Thailand', 'Vietnam',
                'Philippines', 'Myanmar', 'Cambodia', 'Laos', 'Brunei')
population = (33940000, 5704000, 273523615, 69799978, 96483981,
                108116615, 54409800, 16249798, 7275556, 437479)
colors = ('#ff9999','#66b3ff','#99ff99','#ffcc99', '#c2c2f0', '#ffb3e6',
            '#ff6666', '#c2f0c2', '#ffcc99', '#c2f0f0')

```

```python
df = pd.DataFrame({
                'Country': countries,
                'Population': population
                })

df.sort_values(by='Population', ascending=False, inplace=True)
```

```python
x_coords = np.arange(len(df))
plt.bar(x_coords, df['Population'], color=colors, tick_label=df['Country'])
plt.xticks(rotation=90)
plt.ylabel('Population (in millions)')
plt.title('Population of ASEAN Countries')
plt.show()
```
