```python
dessert_sales = {
    'Lava Cake': 14,
    'Mousse': 5,
    'Chocolate Cake': 12,
    'Ice Cream': 19,
    'Truffles': 6,
    'Brownies': 8,
    'Chocolate Chip Cookies': 12,
    'Chocolate Pudding': 9,
    'Souffle': 10,
    'Chocolate Cheesecake': 17,
    'Chocolate Chips': 2,
    'Fugde': 9,
    'Mochi': 13,
}
```

```python
# Mengimpor modul matplotlib.pyplot
import matplotlib.pyplot as plt

# Menyortir penjualan dessert berdasarkan angka penjualan
sorted_desserts = sorted(dessert_sales.items(), key=lambda x: x[1])

# Mengambil nama dessert dan angka penjualan
names, sales = zip(*sorted_desserts)

# Membuat bar chart
plt.figure(figsize=(12, 8))
bars = plt.barh(names, sales, color='#6CABDD')
plt.xlabel('Angka Penjualan')
plt.ylabel('Nama Dessert')
plt.title('Penjualan Dessert pada Minggu Lalu')

# Meng-highlight 3 dessert yang kurang laku
for bar in bars[:3]:
    bar.set_color('#EC3325')

# Menambahkan angka penjualan pada setiap bar
for bar, value in zip(bars, sales):
    plt.text(value, bar.get_y() + bar.get_height() / 2, str(value), va='center', ha='right')

# Menambahkankan watermark
plt.text(0.99, 0.01, 'Made by Faisal A F Rahman (2130511027)', color='gray', fontsize=10, ha='right', transform=plt.gca().transAxes)

plt.tight_layout()
plt.show()
```

```python
# Bar chart tanpa 3 dessert yang kurang laku

# Membuat bar chart
plt.figure(figsize=(12, 8))
bars = plt.barh(names[3:], sales[3:], color='#6CABDD')
plt.xlabel('Angka Penjualan')
plt.ylabel('Nama Dessert')
plt.title('Penjualan Dessert pada Minggu Lalu')

# Menambahkan angka penjualan pada setiap bar
for index, value in enumerate(sales[3:]):
    plt.text(value, index, str(value), va='center', ha='right')

# Menambahkankan watermark
plt.text(0.99, 0.01, 'Made by Faisal A F Rahman (2130511027)', color='gray', fontsize=10, ha='right', transform=plt.gca().transAxes)

plt.tight_layout()
plt.show()
```

Alasan saya memilih menggunakan bar chart secara horizontal adalah karena setiap *dessert* memiliki nama yang cukup panjang, sehingga lebih mudah dibaca dan menghindari tumpang tindih antar nama *dessert*. Selain itu, penggunaan bar chart cocok untuk membandingkan data kategori dalam kasus ini adalah nama *dessert*.

Pada kasus ini, saya menggunakan warna yang berbeda untuk membedakan antara *dessert* yang layak dipertahankan dan yang harus dihapus dari daftar menu.
