```python
# Import pandas and numpy
import pandas as pd
import numpy as np
```

```python
path = "dataset/dataset.csv" # Sesuaikan dengan lokasi file dataset.csv
df_epl = pd.read_csv(path)

apakah_ada_nilai_null_df = df_epl.isnull().values.any()

# Mengecek apakah terdapat nilai null di kolom tertentu, contohnya kolom "High"
apakah_ada_nilai_null_high = df_epl["goals"].isnull().values.any()

# Mencetak hasil pengecekan
print(f"Apakah terdapat nilai null di dataframe? {apakah_ada_nilai_null_df}")
print(f"Apakah terdapat nilai null di kolom Goals? {apakah_ada_nilai_null_high}")
```
